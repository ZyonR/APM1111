1. Table 1 shows a frequency distribution of grades on a final examination in
college algebra. Find the quartiles of the distribution.

```python
import pandas as pd
import numpy as np
import itertools
```

```python
table_1 = {
    "Grade":["90-100","80-89","70-79","60-69","50-59","40-49","30-39"],
    "Number of Students": [9,32,43,21,11,3,1]
}
table_1 = pd.DataFrame(table_1)
table_1[["Lower Limit Grade","Upper Limit Grade"]]= table_1["Grade"].str.split("-",expand = True)
table_1.drop(columns=['Grade'],inplace = True)
table_1 = table_1.iloc[:,[1,2,0]]
table_1["Lower Limit Grade"].astype(int)
table_1["Upper Limit Grade"].astype(int)
table_1
```

We have 4 quartiles, which are Q_1, Q_2, Q_3, Q_4. Which are defined as the follwing:
1. Q_1 - 25% of the data
2. Q_2 - 50% of the data
3. Q_3 - 75% of the data
4. Q_4 - 90% of the data

Let us first find the cumulative frequency of the given grouped data.

```python
table_1["Cumulative Freq."] = table_1["Number of Students"].cumsum()
table_1
```

Now given that te total is 120 students we can use this to determine the quartiles where each quartile shall be represented by the following equation:
1. Q_1 = 120 * 1/4
2. Q_2 = 120 * 1/2
3. Q_3 = 120 * 2/3
4. Q_4 = 120 * 9/10

```python
quartiles = []
proportions = [1/4,1/2,3/4,9/10]
total_n_students = 120
for prop in proportions:
    quartiles.append(total_n_students*prop)
for i,q in enumerate(quartiles):
    print(f"Q_{i+1} = {q}")
```

Given these intervals it shows the following class intervals for each quarters:
1. Q_1 => 30-39
2. Q_2 => 60-69
3. Q_3 => 90-100	
4. Q_4 => No Class Interval Higher than 108

2. On a final examination in statistics, the mean grade of a group of 150
students was 78 and the standard deviation was 8.0. In algebra, however, the mean final
grade of the group was 73 and the standard deviation was 7.6. In which subject was
there the greater (a) absolute dispersion and (b) relative dispersion?

For absolute dispersion we can look at both the standard deviations of the two subject examinations, for relative dispersion, since there are different N for both subject which would entail that there are different means.

```python
statistics = {
    "mean":78,
    "stdDev":8.0,
}
algebra = {
    "mean":73,
    "stdDev":7.6,
}
statistics["coefVar"] = statistics["stdDev"]/statistics["mean"]
algebra["coefVar"] = algebra["stdDev"]/algebra["mean"]

absoluteDispersion = "statistics" if statistics["stdDev"]>algebra["stdDev"] else "algebra"
relativeDispersion = "statistics" if statistics["coefVar"]>algebra["coefVar"] else "algebra"

absoluteDispersionValue = statistics["stdDev"] if statistics["stdDev"]>algebra["stdDev"] else algebra["stdDev"]
relativeDispersionValue = statistics["coefVar"] if statistics["coefVar"]>algebra["coefVar"] else algebra["coefVar"]

print(f"Greater Absolute Dispersion: {absoluteDispersion}\nwith a value of: {absoluteDispersionValue}")
print(f"\nGreater Relative Dispersion: {relativeDispersion}\nth a value of: {relativeDispersionValue}")
```

3.  Prove that the mean and standard deviation of a set of standard scores are
equal to 0 and 1, respectively. Use the following problem to illustrate this: Convert the
set 6, 2, 8, 7, 5 into standard scores.

```python
setOfNums = [6, 2, 8, 7, 5 ]
# to convert to standard scores we must find the z-score where it is given by the formula x-mean/stdDev
meanOfSetOfNums = np.mean(setOfNums)
stdDevOfSetOfNums = np.std(setOfNums)
setOfStandardScores = (np.array(setOfNums) - meanOfSetOfNums)/stdDevOfSetOfNums
print(f"Set of Standard Scores: {setOfStandardScores}")
print(f"Mean of the Set of Standard Scores: {round(np.mean(setOfStandardScores))}")
print(f"Standard Deviation of the Set of Standard Scores: {round(np.std(setOfStandardScores))}")
```

4. Three masses are measured as 20.48, 35.97, and 62.34 g, with standard

deviation of the sum of the masses.

standard deviations are the value of the deviations from the mean for each given x.

```python
masses = [20.48, 35.97, 62.34]
standardDevPerX = [0.21,0.46,0.54]
mean_masses = np.mean(masses)
standardDev = ((sum(np.array(standardDevPerX)**2))/len(standardDevPerX))**(1/2)
print(f"Mean of the Masses: {mean_masses}\nStandard deviation of the sum of the masses: {standardDev}")
```
