```python
import random
def getRandomSample(array,index):
    n_Sample = 4
    sample_Index = f"Sample {index+1}"
    
    take_Sample = random.choices(array, k=n_Sample)
    count_Marcos = take_Sample.count('Marcos')
    count_Robredo = take_Sample.count('Robredo')
    
    proportion_Marcos = count_Marcos/n_Sample
    proportion_Robredo = count_Robredo/n_Sample
    
    return {
        "sampleIndex": sample_Index,
        "sample": take_Sample,
        "propMarcos":proportion_Marcos,
        "propRobredo":proportion_Robredo
    }

electoralCandidatesPopuVotes = ["Marcos","Marcos","Marcos","Marcos","Robredo","Robredo"]
all_SampleData = []
for i in range(15):
    all_SampleData.append(getRandomSample(electoralCandidatesPopuVotes,i))
print(all_SampleData)
```

```python
#transformation of data
sampleIndices = []
voter_1 = [] 
voter_2 = []
voter_3 = []
voter_4 = []
propMarcos = []
propRobredo = [] 

for data in all_SampleData:
    sampleIndices.append(data["sampleIndex"])
    voter_1.append(data["sample"][0])
    voter_2.append(data["sample"][1])
    voter_3.append(data["sample"][2])
    voter_4.append(data["sample"][3])
    propMarcos.append(data["propMarcos"])
    propRobredo.append(data["propRobredo"])

dataframe = {
    "sampleIndices": sampleIndices,
    "voter_1" : voter_1, 
    "voter_2" : voter_2,
    "voter_3" : voter_3,
    "voter_4" : voter_4,
    "propMarcos" : propMarcos,
    "propRobredo" : propRobredo 
}

import pandas as pd

df = pd.DataFrame(dataframe)
df
```

```python
# get average prop of sample data
popuProp_Marcos = round (electoralCandidatesPopuVotes.count('Marcos')/len(electoralCandidatesPopuVotes),2)
popuProp_Robredo = round (electoralCandidatesPopuVotes.count('Robredo')/len(electoralCandidatesPopuVotes),2)

print(f"Population Proportion of Marcos: {popuProp_Marcos}\nPopulation Proportion of Robredo: {popuProp_Robredo}")

print(f"\n\nAverage Sample Proportion of Marcos: {df['propMarcos'].mean()}\nAverage Sample Proportion of Robredo: {df['propRobredo'].mean()}")
```
